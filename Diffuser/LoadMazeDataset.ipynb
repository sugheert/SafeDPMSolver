```python
import torch
import matplotlib.pyplot as plt
```

```python
import sys,os
sys.path.append(os.path.join(os.path.abspath(os.getcwd())))
```

```python
from maze_dataset import MazeDataset, render_maze_ax

ds = MazeDataset(os.path.join(os.getcwd(), "data/umaze_v2"))
# or xy_only=True for [N, 128, 2]

dl = torch.utils.data.DataLoader(ds, batch_size=64, shuffle=True)
batch = next(iter(dl))

print(batch.shape)           

world = ds.unnormalize(batch)    # back to world coords for plotting

fig, ax = plt.subplots()
render_maze_ax(ax, ds.maze_map, ds.cell_size)
ax.plot(world[0, :, 0], world[0, :, 1])   # single trajectory
```
